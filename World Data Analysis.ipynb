```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('f5791aee-555c-4113-b058-ea8a3dd6569a_Data.csv')
```

```python
df.head()
```

```python
# NULL VALUES

df.isnull().sum()

# Less than 5% null values were found in all the columns so dropping them for ease of analysing data.


# Removing null values

df.dropna(inplace = True)
```

```python
# Checking for successfull removal of null values with percentage of null values
 
df.isnull().sum()/len(df)
```

```python
# CHECKING FOR DUPLICATES

df.duplicated().sum()

# Since no duplicates are present, we can go ahead to start visualising the data
```

# DATA VISUALISATION

```python
df['Series Name'].unique()
```

```python
# ANNUAL GDP GROWTH

df3 = df[df['Series Name']=='GDP growth (annual %)']
df3
```

```python
color = ['#96ceb4', '#ff6f69','#ffeead']

df3.groupby('Country Name')['2013 [YR2013]'].mean().sort_values(ascending = False).plot(kind = 'bar', color = color )
plt.title('Annual GDP growth percentage in 2013')
plt.ylabel('GDP growth')
plt.show()


# INSIGHTS

1. Turkey, China and India had the best GDP growth 
2. Spain and Italy ranked in the negattive
```

# FOR 2022

```python
color = ['#96ceb4', '#ff6f69','#ffeead']

df3.groupby('Country Name')['2022 [YR2022]'].mean().sort_values(ascending = False).plot(kind = 'bar', color = color )
plt.title('Annual GDP growth percentage in 2022')
plt.ylabel('GDP growth')
plt.show()


# INSIGHTS

1. Saudi Arabia and India have the top spot in GDP in 2022 whilst, Russian federation is ranking at last
```

# Population density people per sq. km of land area

```python
df4 = df[df['Series Name']=='Population density (people per sq. km of land area)']
df4
```

```python
color = ['#96ceb4', '#ff6f69','#ffeead']

df4.groupby('Country Name')['2016 [YR2016]'].mean().sort_values(ascending = False).plot(kind = 'bar', color = color )
plt.title('Population density 2016')
plt.ylabel('Population density')
plt.show()


# INSIGHTS

# Korea and Netherlands had the most population density in 2021
```

```python
color = ['#96ceb4', '#ff6f69','#ffeead']

df4.groupby('Country Name')['2013 [YR2013]'].mean().sort_values(ascending = False).plot(kind = 'bar', color = color )
plt.title('Population density 2013')
plt.ylabel('Population density')
plt.show()


# INSIGHTS

# Korea and Netherlands had the most population density in 2021
```

```python
# In terms of population density, there has been no significatn changes in each country from 2013 -2021. 
# There has been a steady rate of population over 9 years 
```

# CO2 emissions (metric tons per capita)

```python
df5 = df[df['Series Name']=='CO2 emissions (metric tons per capita)']
df5
```

```python
color = ['#be77fb', '#3eecac']

df5.groupby('Country Name')['2013 [YR2013]'].mean().sort_values(ascending = False).plot(kind = 'bar', color = color )
plt.title('CO2 Emissions 2013')
plt.ylabel('CO2 Emission')
plt.show()


# INSIGHTS
# Australia, US  and Saudi Arabia had the highest CO2 emissions 
# Interestingly CHina and India though has the highest population ranked much lower with India being the lowest
# in the C02 Emissions in 2013
```

# Fertility rate, total (births per woman)

```python
df6 = df[df['Series Name']=='Fertility rate, total (births per woman)']
df6
```

```python
color = ['#be77fb', '#3eecac']
df6.groupby('Country Name')['2013 [YR2013]'].mean().sort_values(ascending = False).plot(kind = 'bar', color = color )
plt.title('Fertility Rate (Total Births) in women')
plt.ylabel('Fertility Rate')
plt.show();


# INSIGHTS

# Saudi Arabia, Indonesia , India , Argentina and Mexico has the highest fertility rates
# Korea has the lowest fertility rate yet the highest population density
```

```python
color = ['#be77fb', '#3eecac']
df6.groupby('Country Name')['2021 [YR2021]'].mean().sort_values(ascending = False).plot(kind = 'bar', color = color )
plt.title('Fertility Rate (Total Births) in women 2021')
plt.ylabel('Fertility Rate')
plt.show();


# INSIGHTS

# Saudi Arabia, Indonesia , India , TURKEY and France has the highest fertility rates
# Korea has the lowest fertility rate yet the highest population density
```
